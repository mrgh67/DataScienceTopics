```python
import numpy as np
import pandas as pd
import sklearn as sk

import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge

from sklearn import cross_validation

from sklearn.cross_validation import KFold
from sklearn.cross_validation import train_test_split
from sklearn.cross_validation import cross_val_score

from sklearn.preprocessing import PolynomialFeatures

from sklearn.pipeline import make_pipeline

from sklearn.datasets import load_diabetes

from sklearn.metrics import mean_squared_error

from itertools import combinations
```

```python
dD = load_diabetes()
```

```python
pDF = pd.DataFrame(dD.data)
pDF.describe()
pDF.columns = ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J"]
pDF.columns
pDF.describe()
predictors = ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J"]
```

Perform and initial loop using LinearRegression(). 
Note that in addition to other parameters of the model the number of folds in the cross validation should also be considered.

```python
cross_validation_errors = {}
training_errors = {}
for i in range(2, len(predictors)):
    for c in combinations(predictors, r=i):
        for n_flds in (2, 4, 6, 8, 10):
            indxList = [p for p in c]
            indxList.append(str(n_flds))
            idxName = '_'.join(indxList)
            X = pDF[list(c)].values
            y = dFNTarget.values
            kfold = KFold(len(pDF), n_folds=n_flds)
            mean_sq_error = []
            tr_mse = []
            for train, test in kfold:
                Xtrain, ytrain, Xtest, ytest = X[train], y[train], X[test], y[test]
                model = LinearRegression()
                model.fit(Xtrain,ytrain)
                ypredtrain = model.predict(Xtrain)
                ypred = model.predict(Xtest)
                mean_sq_error.append(mean_squared_error(ypred, ytest))
                tr_mse.append(mean_squared_error(ypredtrain, ytrain))
            cross_validation_errors[idxName] = np.mean(mean_sq_error)
            training_errors[idxName] = np.mean(tr_mse)
```

I added another dictionary to capture the training MSE as well.

```python
#These are the results for the test set
svedF = pd.DataFrame.from_dict(cross_validation_errors, orient='index')
svedF.columns = ['MSE']
svedF = svedF.sort('MSE', ascending=True)
svedF.head()
```

```python
svedFtrain = pd.DataFrame.from_dict(training_errors, orient='index')
svedFtrain.columns = ['MSE']
svedFtrain = svedFtrain.sort('MSE', ascending=True)
svedFtrain.head()
```

Now construct multiple iterative loops to try and zero in on a good predictor. 
This loop over the following parameters: all combinations of predictors, degree of poylynomial from 2 to 9, alpha from 0.1 to 0.0001 in 0.05 increments. I took the above result and used number of folds = 4 as the number of folds to use in cross validation.

I chose Ridge as my regression algorithm.

NB: I have removed the loop ranges because of processing time. The optimal parameters remain.

```python
cross_validation_errors2 = {}
for i in range(2, len(predictors)):
    #print i
    for degree in [2]:
        #print "%d " % degree
        for a in [0.01]:
            #print a
            for c in combinations(predictors, r=i):
                #print degree
                indxList = [p for p in c]
                indxList.append(str(a))
                indxList.append(str(degree))
                #print indxList
                idxName = '_'.join(indxList)
                #print idxName
                #j=[p for p in c]
                #print j
                X = pDF[list(c)].values
                y = dFNTarget.values
                kfold = KFold(len(pDF), n_folds=4)
                mean_sq_error = []
                for train, test in kfold:
                    Xtrain, ytrain, Xtest, ytest = X[train], y[train], X[test], y[test]
                    model = make_pipeline(PolynomialFeatures(degree), Ridge(alpha=a))
                    model.fit(Xtrain,ytrain)
                    ypred = model.predict(Xtest)
                    mean_sq_error.append(mean_squared_error(ypred, ytest))
            cross_validation_errors2[idxName] = np.mean(mean_sq_error)
```

```python
svedF2 = pd.DataFrame.from_dict(cross_validation_errors2, orient='index')
svedF2.columns = ['MSE']
svedF2 = svedF2.sort('MSE', ascending=True)
svedF2
```

```python
svedF2 = svedF2.sort('MSE', ascending=False)
plt.figure(figsize=(10, 5))
ax = plt.gca()
ax.plot(svedF2, '-',color="blue")
plt.title("MSE Test Set")
plt.ylabel("MSE")
plt.xlabel("Iteration")
ax.set_xlim((0,8))
```

One of the things that slightly concerns me is that this approach, while resulting in a possibly good predictor with low MSE on the test set, doesn't really provide information about the model itself. So take this data for example. Trying to predict quality of life from physiological measurements for Diabetics is important, and understanding the role of each of the predictors is almost certainly going be important. I suppose the magnitude of the coefficients would indicate the relative importance of each of the predictors. I do, however, prefer the overall strategy of trying to determine which predictors are important and therefore should be included in the model in the first place. The information reported by the "stats model" ols regression models is appealing.

So I did some further exploring with this model.

I also quite like the flexibility offered by being able to enter a formula.
